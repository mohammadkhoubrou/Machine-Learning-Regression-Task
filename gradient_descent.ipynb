```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def lin(x,w,b):
    return w * x + b * np.ones(x.shape)
```

```python
x_train = np.array([1,2,3,2,4,5,8,6,9,7])
y_train = np.array([9,8,6,9,7,3,3,4,3,2])
```

```python
plt.scatter(x_train,y_train,c='r',label="data")
plt.xlabel("x_train")
plt.ylabel("y_train")
plt.show()
```

```python
w,b = -1.3,12
prediction = lin(x_train,w,b)
prediction
```

```python
plt.plot(x_train, prediction, c='b',label= "prediction")
plt.scatter(x_train,y_train, c='r', label="data")
plt.xlabel = "x"
plt.ylabel = "y"
plt.title = "prediction on data"
plt.show()
```

```python
def mse(prediction, target):
    p = 0
    for i in range(len(prediction)):
        p += (prediction[i]-target[i])**2
    return p/(2*len(prediction))
```

```python
def deriv(prediction, target,x):
    dw = 0
    db = 0
    for j in range(0,len(target)):
        dw += (prediction[j]-target[j])*x[j]
        db += (prediction[j]-target[j])
    return dw/len(prediction) , db/len(prediction)
```

```python
def gd(x,y, w_in,b_in,cost,gradient,alpha, iterations,lin):
    w = w_in
    b = b_in
    j = [] 
    p = []
    # pred = lin(x, w, b)
    for i in range(iterations):
        pred = lin(x, w, b)
        dj_w, dj_b = gradient(pred, y, x)
        w = w - alpha * dj_w
        b = b - alpha * dj_b
        if i < 100000:
            j.append(cost((lin(x, w, b)), y))
            p.append([w,b])
    return w, b, j, p
```

```python
w_in,b_in = 0,0
iterations = 10000
alpha = 1.0e-2
w,b,cost,par = gd(x_train,y_train,w_in, b_in, mse, deriv,alpha,iterations,lin)
cost
```
